```python
static_bike_station_strcuture = bos_station_info['data']['stations'][0].keys()

static_bike_station_strcuture
```

```python
import json

raw_taipei_bike_station_data = json.load(open('raw_new_taipei_bike_station_data.json'))

raw_taipei_bike_station_data
```

```python
res = []

for tpe_station_static in raw_taipei_bike_station_data:
    tmp = {}
    # SrcUpdateTime = tpe_station['SrcUpdateTime']
    StationUID = tpe_station_static['StationUID']
    name = tpe_station_static['StationName']['En']
    geo_hash = tpe_station_static['StationPosition']['GeoHash']
    lon = tpe_station_static['StationPosition']['PositionLon']
    lat = tpe_station_static['StationPosition']['PositionLat']
    for s in static_bike_station_strcuture:
        if s == 'name':
            tmp[s] = name
        if s == 'station_type':
            pass
        if s == 'short_name':
            tmp[s] = geo_hash
        if s == 'lon':
            tmp[s] = lon
        if s == 'rental_methods':
            tmp[s] = ["KEY", "CREDITCARD"]
        if s == 'lat':
            tmp[s] = lat
        if s == 'station_id':
            tmp[s] = StationUID
        # if s == 'electric_bike_surcharge_waiver':
        #     tmp[s] = false
        if s == 'capacity':
            tmp[s] = 10
        


    res.append(tmp)

tpe_stations_static = bos_station_info

tpe_stations_static['data']['stations'] = res

# the json file where the output must be stored 
out_file = open("new_tpe_station_static.json", "w") 
  
json.dump(tpe_stations_static, out_file) 
  
out_file.close()
```

```python
from datetime import datetime, timezone, timedelta


res = []

def timestamp_to_unixtime(dt_str: str):
    # Convert string to datetime object
    dt_obj = datetime.strptime(dt_str, "%Y-%m-%dT%H:%M:%S%z")
    
    # Convert to Unix time (seconds since 01-01-1970 00:00:00 UTC)
    unix_time = dt_obj.timestamp()
    
    return unix_time


for tpe_station in raw_taipei_bike_status_data:
    tmp = {}
    SrcUpdateTime = tpe_station['SrcUpdateTime']
    StationUID = tpe_station['StationUID']
    for s in structure:
        if s == 'num_docks_disabled':
            tmp[s] = 0
        if s == 'last_reported':
            tmp[s] = timestamp_to_unixtime(SrcUpdateTime)
        if s == 'num_scooters_available':
            tmp[s] = 0
        if s == 'num_scooters_unavailable':
            tmp[s] = 0
        if s == 'is_renting':
            tmp[s] = 1
        # if s == 'eightd_has_available_keys':
        #     tmp[s] = 1
        if s == 'num_bikes_disabled':
            tmp[s] = 0
        if s == 'station_id':
            tmp[s] = StationUID
        if s == 'is_returning':
            tmp[s] = 1
        if s == 'num_bikes_available':
            tmp[s] = 10
        if s == 'num_ebikes_available':
            tmp[s] = 0
        if s == 'num_docks_available':
            tmp[s] = 10
        if s == 'is_installed':
            tmp[s] = 1


    res.append(tmp)
            
tpe_stations = bos_station_status

tpe_stations['data']['stations'] = res

# the json file where the output must be stored 
out_file = open("new_tpe_status.json", "w") 
  
json.dump(tpe_stations, out_file) 
  
out_file.close()
        

 # {'legacy_id': '6',
 #  'num_docks_disabled': 0,
 #  'last_reported': 1709018921,
 #  'num_scooters_available': 0,
 #  'num_scooters_unavailable': 0,
 #  'is_renting': 1,
 #  'eightd_has_available_keys': False,
 #  'num_bikes_disabled': 1,
 #  'station_id': 'f83466c6-0de8-11e7-991c-3863bb43a7d0',
 #  'is_returning': 1,
 #  'num_bikes_available': 7,
 #  'num_ebikes_available': 0,
 #  'num_docks_available': 6,
 #  'is_installed': 1}]

```
