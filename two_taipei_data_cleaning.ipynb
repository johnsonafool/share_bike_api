```python
import pandas as pd
```

```python
calendar_df = pd.read_csv('/home/csl/otp/two_taipei_gtfs/calendar.txt')

calendar_df
```

```python
# sanize chekcing
calendar_df.drop_duplicates()

calendar_df
```

```python
# filter that only on two taipeis
# Define the patterns
patterns = ["TPE", "NWT", "NTMC", "TRTC"]

# Create a boolean mask with OR conditions for multiple patterns
mask = calendar_df['service_id'].str.startswith(patterns[0])
for pattern in patterns[1:]:
    mask |= calendar_df['service_id'].str.startswith(pattern)

# Filter rows
filtered_calendar_df = calendar_df[mask]

# Show the filtered DataFrame
filtered_calendar_df.to_csv('/home/csl/otp/two_taipei_gtfs/calendar.txt', index=False)

filtered_calendar_df
```
